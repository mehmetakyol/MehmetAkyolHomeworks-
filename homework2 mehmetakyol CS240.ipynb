```python
pres = nsfg.ReadFemResp()
pres
```

```python
fmar1age = pres['fmar1age']
fmar1age
```

```python
import thinkstats2
hist=thinkstats2.Hist(fmar1age)
hist
```

```python
for val in sorted(hist.Values()):
    print ( val,hist[val])
```

```python
import thinkplot
thinkplot.Hist(hist)
thinkplot.Config(xlabel='ages',ylabel='frequency')
```
